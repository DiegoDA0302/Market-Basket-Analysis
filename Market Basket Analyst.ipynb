```python
# The first step is to import the libraries required to conduct a detailed analysis.
import pandas as pd
```

```python
# I will now load the .csv file and use .head to take a preliminary look at the data.
df = pd.read_csv(r'C:\Users\diego\OneDrive\Documents\DataScient\Mi Primera Experiencia\tickets.csv')
df.head()
```

```python
# Description of the data contained in the dataset.
df.info()
```

```python
# I convert the data type of the 'fecha' column from object to date.
df['fecha'] = pd.to_datetime(df['fecha'])
df.info()
```

```python
# Now I will only take the 'id_pedido' and 'nombre_producto' columns to perform Market Basket Analytics.
df_basket = df[['id_pedido','nombre_producto']]
df_basket
```

```python
# I perform a grouping by 'id_pedido' to find out which products are in each order.
df_agrupado = df_basket.groupby('id_pedido')['nombre_producto'].apply(lambda producto: ','.join(producto))
df_agrupado
```

```python
# I convert the dataset into a binary system to identify in a table the purchases of each product.
df_transacciones = df_agrupado.str.get_dummies(sep=',')
df_transacciones
```

```python
# Here, I make the first calculation, which is the support variable, where it shows the probability of buying item x out of the total purchases.
support = df_transacciones.mean() * 100
support.sort_values(ascending=False)
```

```python
# Function to calculate the confidence between two products in the sample.
def confidence(antecedent, consequent):
    # Cases where both products were purchased.
    set_ac = df_transacciones[(df_transacciones[antecedent] == 1) &
                                   (df_transacciones[consequent] == 1)]
    # Confidence = joint purchases / purchases of product A.
    return len(set_ac) / df_transacciones[antecedent].sum()
```

```python
# Function to calculate the lift between two products in the sample.
def lift(antecedent, consequent):
    support_a = df_transacciones[antecedent].mean()
    support_c = df_transacciones[consequent].mean()
    count_ac = len(df_transacciones[(df_transacciones[antecedent] == 1) &
                                   (df_transacciones[consequent] == 1)])
    support_ac = count_ac / len(df_transacciones)
    return support_ac / (support_a * support_c)
```

```python
from itertools import combinations

# Define a threshold for the minimum confidence.
threshold_confidence = 0.05
associations = []

# Generate product combinations and calculate confidence and lift.
for antecedent, consequent in combinations(df_transacciones.columns, 2):

    # Support of the antecedent
    support_a = df_transacciones[antecedent].mean()

    # Calculate confidence
    conf = confidence(antecedent, consequent)
    if conf > threshold_confidence:
        associations.append({
            'antecedent': antecedent,
            'consequent': consequent,
            'support_a': round(support_a * 100,1),
            'confidence': round(conf * 100,1),
            'lift': round(lift(antecedent, consequent),1)
        })


# Convert the associations into a DataFrame.
df_associations = pd.DataFrame(associations)

# Sort the associations by confidence from highest to lowest.
df_associations.sort_values(by='lift', ascending=False, inplace=True)
df_associations
```

```python
# Create a table with the unique products and the corresponding columns.
df_unic_products = df[['id_producto', 'id_seccion', 'id_departamento', 'nombre_producto']].drop_duplicates()
df_unic_products
```

```python
# In this DataFrame, I merge the two tables I created to identify the products that are related in each purchase.
df_associations_enriched = df_associations.merge(df_unic_products, left_on='antecedent', right_on='nombre_producto', how='left').drop(columns=['nombre_producto'])
df_associations_enriched.columns = ['antecedent', 'consequent', 'support_a', 'confidence', 'lift', 'id_producto_a', 'id_seccion_a', 'id_departamento_a']
df_associations_enriched
```

```python
df_associations_enriched.to_csv('reglas.csv', index=False, sep=';', decimal=',')
```
